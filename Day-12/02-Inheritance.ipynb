Inheritance in Python is a fundamental concept in Object-Oriented Programming (OOP) that allows a new class (called the child or derived class) to inherit attributes and methods from an existing class (called the parent or base class). This promotes code reusability and establishes an "is-a" relationship between classes. 

## Single inheritance 

In Python is a type of inheritance where a child class (also known as a subclass or derived class) inherits attributes and methods from only one parent class (also known as a superclass or base class). This creates a direct, one-to-one relationship between a parent and its child.

```python
class Animal:
    def __init__ (self, name, age):
        self.name = name
        self.age = age
        
    def into(self):
        print(f"My name is {self.name}")
```

```python
class Cat(Animal):
    pass
```

```python
cat = Cat()
```

```python
class Cat(Animal):
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
        
    def details(self):
        print(f"{self.name}, {self.age}, {self.color}")
        
    def sound(self):
        print("meow")
```

```python
cat = Cat("Simba", 3, "Golden")
```

```python
cat.into()
```

```python
cat.details()
```
