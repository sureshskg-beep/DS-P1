# Key characteristics of OOP

Encapsulation: The practice of bundling data (attributes) and the methods (functions) that operate on that data into a single unit called a class. It protects an object's internal state from outside interference by making its data private and only allowing access through public methods.
Inheritance: A mechanism that allows a new class (a child class) to inherit properties and behaviors from an existing class (a parent class). This promotes code reusability and establishes a hierarchical relationship between classes.
Polymorphism: The ability of a single interface to represent different types of objects or actions. It allows an object to take on many forms and enables functions or methods to perform different actions depending on the object they are operating on.
Abstraction: The process of simplifying complex systems by modeling classes based on their essential features and hiding unnecessary details. It focuses on what an object needs to do, rather than how it does it.
Classes and objects: OOP organizes code into objects, which represent real-world entities, and classes, which serve as blueprints for creating objects.
Message Passing: Objects interact with each other by sending messages (often in the form of method calls) to perform actions.
