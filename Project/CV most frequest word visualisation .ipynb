```python
# Import libraries
from docx import Document
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
import string

# Download stopwords (run once)
nltk.download("stopwords")

# -------------------------------
# Step 1: Load resume (.docx)
# -------------------------------
doc = Document("C:\\Users\\Suresh\\Desktop\\DSML\\Projects\\Resume-SURESH-SK Final.docx")


text = ""
for para in doc.paragraphs:
    text += para.text + " "

# -------------------------------
# Step 2: Text cleaning
# -------------------------------
text = text.lower()  # lowercase
text = text.translate(str.maketrans("", "", string.punctuation))  # remove punctuation

stop_words = set(stopwords.words("english"))
custom_stopwords = {
    "years", "experience", "managed", "management", "including",
    "across", "support", "worked", "project", "program"
}

words = [
    word for word in text.split()
    if word not in stop_words and word not in custom_stopwords and len(word) > 2
]

clean_text = " ".join(words)

# -------------------------------
# Step 3: Create WordCloud
# -------------------------------
wordcloud = WordCloud(
    width=800,
    height=400,
    max_words=100,
    background_color="white",
    colormap="winter"
).generate(clean_text)

# -------------------------------
# Step 4: Visualize
# -------------------------------
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Frequent Words in Resume", fontsize=14)
plt.show()
```
