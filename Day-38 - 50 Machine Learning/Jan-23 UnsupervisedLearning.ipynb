# Unsupervised Learning

### Definition

Unsupervised learning is a class of machine learning techniques where the model is trained on unlabeled dataâ€”that is, the dataset contains no target variable (y). The objective is to discover hidden patterns, structures, or relationships within the data.

In short: the algorithm learns from the data itself, without knowing the correct answers.

### Key characteristics

No predefined output labels

Focus on structure discovery, not prediction

Often used for exploration, preprocessing, and compression
