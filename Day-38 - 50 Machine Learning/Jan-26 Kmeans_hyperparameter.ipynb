```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Cancer_RNA/data.csv", index_col=0)
labels = pd.read_csv("Cancer_RNA/labels.csv", index_col=0)
```

```python
data.head()
```

```python
data.shape
```

```python
scaler = StandardScaler()
X  = scaler.fit_transform(data)
```

```python
X.shape
```

# HyperParameters

```python
scores = []
for k in range(2, 10):
    model = KMeans(n_clusters=k, random_state=99)
    model.fit(X)
    
    scores.append(model.inertia_)
    del model
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12,4))

k_range = range(1, len(scores) + 1)
plt.plot(k_range, scores)
plt.scatter(k_range, scores, s=50, marker="*", c="red")

plt.xlabel("Number of Clusters (k)")
plt.ylabel("Score")
plt.show()
```
