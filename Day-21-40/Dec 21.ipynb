## Regular Expresion in Python

Regular Expressions (often called regex) are patterns used to search, match, extract, or replace text.
In Python, regex is handled using the built-in re module.

Think of regex as smart text searching.

Why do we need Regular Expressions?
You use regex when:
You want to find patterns, not exact words
Validate input (email, phone number, password)
Extract useful data from text
Replace multiple text patterns at once

Example:
Find all email addresses in a file
Check if a phone number is valid
Extract dates from a document
