Numpy - Slicing 

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4, 5, 6,])
```

```python
print(a[-3])
```

```python
a[2::1]
```

```python
#2D

arr = np.random.randint(0, 10, size = (5, 4))
arr
```

```python
arr[1, 0] #or arr [1][0]
```

```python
arr [-1, -1]
```

```python
#slicing 2D array : (aray[<row_slice>, <col_slice>])
arr[1:3, 0:3]
```

```python
arr[:2,3]
```

```python
#3D
arr = np.random.randint(0, 10, size = (3,4,3))
arr
```

```python
arr[1, -1, 2]
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr[2]
```

```python
arr[:,:2,:]
```
