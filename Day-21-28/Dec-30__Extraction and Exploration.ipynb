```python
titanic_df[
    (titanic_df.Embarked == "S") & (titanic_df["Sex"] == "male")
].value_counts('Survived')
```

```python
pd.crosstab(
    titanic_df.Sex,
    titanic_df.Survived,
    margins = True
)
```

```python
pd.crosstab(
    titanic_df.Sex,
    titanic_df.Survived,
    margins = False
)
```

```python
pd.crosstab(
    titanic_df.Sex,
    titanic_df.Survived,
    margins = True,
    colnames = ["Alive"]
)
```

```python
import numpy as np

pd.crosstab(
    titanic_df.Sex,
    titanic_df.Survived,
    margins = True,
    values = titanic_df.Fare,
    aggfunc = "mean"
)
```

```python
pd.crosstab(
    titanic_df.Sex,
    titanic_df.Survived,
    margins = True,
    margins_name= "Total"
)
```

```python
pd.crosstab(
    [titanic_df.Embarked, titanic_df.Pclass],
    [titanic_df.Sex, titanic_df.Survived],
    margins = True
)
```

```python
pd.crosstab(
    [titanic_df.Embarked, titanic_df.Pclass],
    [titanic_df.Sex, titanic_df.Survived],
    margins = True
).T
```
