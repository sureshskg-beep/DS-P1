# data frame 

A pandas DataFrame uses a variety of data types, or dtypes, to store information efficiently. These primarily include numeric types, text (string), boolean, and date/time types, many of which are built on NumPy data types. 
The most common pandas data types are:
object: Used for text (strings) or when a column contains mixed data types.
int64: For integer numbers (whole numbers without decimals). Pandas defaults to 64 bits for performance.
float64: For floating-point numbers (numbers with decimals).
bool: For Boolean values (True and False).
datetime64: For date and time values, which allows for time-series operations.
timedelta[ns]: For differences between two datetimes.
category: For a finite list of text values (e.g., "small", "medium", "large"). This is more memory-efficient for repetitive data.
StringDtype: A dedicated extension type for string data that handles missing values better than the default object dtype.
