```python
# Fibonacci sequence generator in Python

def fibonacci(n):
    """
    Generate the first n terms of the Fibonacci sequence.
    :param n: Number of terms (non-negative integer)
    :return: List containing the Fibonacci sequence
    """
    if n <= 0:
        return []  # No terms for zero or negative input
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    sequence = [0, 1]
    for _ in range(2, n):
        sequence.append(sequence[-1] + sequence[-2])
    return sequence


# Main program
try:
    terms = int(input("Enter the number of Fibonacci terms: "))
    if terms < 0:
        print("Please enter a non-negative integer.")
    else:
        result = fibonacci(terms)
        print(f"Fibonacci sequence ({terms} terms): {result}")
except ValueError:
    print("Invalid input. Please enter an integer.")
How it works:
Function fibonacci(n):

Returns an empty list if n <= 0.
Handles small cases (n = 1 or n = 2) directly.
Uses a loop to generate the sequence for larger n.
Input Validation:

Ensures the user enters a valid integer.
Rejects negative numbers.
Example Runs:


Copy code
Enter the number of Fibonacci terms: 7
Fibonacci sequence (7 terms): [0, 1, 1, 2, 3, 5, 8]
If you want, I can also give you a recursive ver
```
