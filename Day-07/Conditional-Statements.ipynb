````python
## find if given string is palindrome or not?
````

````python
value = "racecar"
````

````python
if value == value[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
````

````python
if (value == value[::-1]):
    print(f"{value} is Palindrome")

else:
    print(f"value is not palindrome")
````
