```python
neighborhood_stats = (
    df_top10pct
    .groupby("Neighborhood")["SalePrice"]
    .agg(mean_saleprice="mean", median_saleprice="median")
    .sort_values(by="mean_saleprice", ascending=False)
)

display(neighborhood_stats.head(10))
```

### **Problem 3: Handling Missing Data and Imputation**  
1. Identify columns with **missing values**.
2. Impute the missing values:
   - For **numerical columns**, replace missing values with the **median**.
   - For **categorical columns**, replace missing values with the **most frequent** value.
3. Justify your choice of imputation methods and check if any missing data remains.

```python
missing_count = house_price.isna().sum()
missing_percent = (missing_count / len(house_price)) * 100

missing_df = (
    pd.DataFrame({
        "Missing Count": missing_count,
        "Missing %": missing_percent
    })
    .query("`Missing Count` > 0")
    .sort_values(by="Missing %", ascending=False)
)

print(f"Total columns with missing values: {missing_df.shape[0]}")
display(missing_df)
```

```python
num_cols2 = house_price.select_dtypes(include=["number"]).columns
cat_cols2 = house_price.select_dtypes(include=["object", "category"]).columns

print("Numerical columns:", len(num_cols2))
print("Categorical columns:", len(cat_cols2))
```

```python
df_imputed = house_price.copy()

# Impute numerical columns with median
for col in num_cols2:
    if df_imputed[col].isna().sum() > 0:
        median_value = df_imputed[col].median()
        df_imputed[col] = df_imputed[col].fillna(median_value)

# Impute categorical columns with mode
for col in cat_cols2:
    if df_imputed[col].isna().sum() > 0:
        mode_value = df_imputed[col].mode()[0]  
        df_imputed[col] = df_imputed[col].fillna(mode_value)

print("Imputation completed.")
```

```python
remaining_missing = df_imputed.isna().sum().sum()

print("Total missing values remaining:", remaining_missing)


remaining_cols = df_imputed.columns[df_imputed.isna().sum() > 0]
print("Columns still having missing values:", list(remaining_cols))

```

```python
print("Missing values BEFORE imputation:", house_price.isna().sum().sum())
print("Missing values AFTER imputation :", df_imputed.isna().sum().sum())
```
