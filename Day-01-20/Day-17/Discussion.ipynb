```python
#__call__

class PassCheck:
    def __init__(self, password : str):
        self.__password = password

    def __call__(self):
        if (
            len(self.__password) >= 8 and
            not self.__password.isalpha()
        ):
            print("Strong Password")
        else:
            print("Weak Password")
```

```python
pass_check = PassCheck("hello23456")
```

```python
pass_check()
```

```python
# if given number is prime

def isPrime(num: int) -> bool:
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

list_numbers = [22,54,78,61,21,17,8]
```

```python
prime_numbers = [num for num in list_numbers if isPrime(num)]
prime_numbers
```

```python
obj = filter(isPrime, list_numbers)
list(obj)
```

```python
# (2024, True),
# (2023, False),
# (1900, False),
# (2000, True),
# (2400, True),
# (2100, False),
# (4, True),
# (2019, False),
# (2021, False),

def check_leap_year(year : int):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 :
                return "Leap Year"
        else:
            return "Leap Year"
    else:
        return "Not Leap Year"
        
```
