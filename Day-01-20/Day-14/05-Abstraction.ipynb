## Abstraction in Python:

A core principle of Object-Oriented Programming (OOP), involves hiding complex implementation details and exposing only the essential functionalities to the user. This simplifies interaction with objects and classes, making code more manageable, readable, and scalable.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass
    
    def volume(self):
        return "Not Applicable for given shape"
```

```python
Shape = Shape()
```

```python
class Rectangle(Shape):
    def __init__(self, width, length):
        self.w = width
        self.l = length
        
    def area(self):
        return self.l*self.w
    
```

```python
rect = Rectangle(50,10)
```

```python
rect.area()
```

```python
class Cube(Shape):
    def __init__(self, l):
        self.l = l
        
    def area(self):
        return 6 * self.l ** 2
    
    def volume(self):
        return self.l ** 3
```
