# Single Inheritance

```python
class Parent1:
    def func1(self):
        print("this is parent 1")
        
class Parent2:
    def func2(self):
        print("this is parent 2")
        
class Parent3:
    def func3(self):
        print("this is parent 3")  
```

```python
class Child(Parent1, Parent2, Parent3):
    def func4(self):
        print("This is child function")
```

```python
obj = Child()
```

```python
obj.func4()
```

```python
obj.func2()
```

```python
obj.func1()
```

# 3. Multi-level inheritance

Multilevel inheritance is an object-oriented programming concept where a class derives from a class that is already inherited from another class, creating a linear chain of inheritance (A -> B -> C). This structure allows a class to inherit attributes and methods from all its ancestors in the chain, promoting code reusability and the creation of specialized classes from existing ones. An example is a Knight class inheriting from a Warrior class, which in turn inherits from a GameCharacter class. 

```python
class A:
    def method_a(self):
        print("This is menthod from class A")
        
class B(A):
    def method_b(self):
        print("This is menthod from class B")
        
class C(B):
    def method_c(self):
        print("This is menthod from class C")
```

```python
obj = C()
```

```python
obj.method_a()
```

```python
obj.method_b()
```

```python
obj.method_c()
```

# Designing a simple hierarchy of employees in a company

- The Person class stores basic details like name and age. 
- The Employee class inherits from Person and adds an employee ID. 
- The Manager class inherits from Employee and adds a department. 

Write these classes and implement a method get_details() in the Manager class that returns a string with name, age, employee ID, and department. 

```python
class Person:
    """
    __docs__
    """
    def __init__(self, name: str = "Suresh SKG", age: int = 40):
        """
        __docs__
        """
        self.name = name
        self.age = age
        
class Employee(Person):
    def __init__(self, name, age, employee_id: str = "107"):
        super().__init__(name, age)
        self.employee_id = employee_id
        
class Manager(Employee):
    def __init__(self, name, age, employee_id, depart: str = "Compliance Team"):
        super().__init__(name, age, employee_id)
        self.depart = depart
        
    def get_details(self):
        return (f"Name: {self.name}, Age: {self.age}, "
                f"Employee ID: {self.employee_id}, Department: {self.depart}")
    
```

## Hierarchical Inheritance



Hierarchical inheritance is an object-oriented programming concept where multiple classes inherit from a single parent (or base) class, creating a tree-like structure. This structure promotes code reuse by allowing all child classes to share the common properties and behaviors of the parent class, while also enabling each child class to have its own unique attributes and methods. 
Key characteristics
Single inheritance source: All derived classes originate from one superclass.
Tree-like structure: The base class is at the top, with multiple child classes branching out below it.
Code reusability: Common features are defined in the parent class, which are then inherited by all children, preventing redundant code.
Extensibility: Each child class can add its own specific functionality without affecting the other child classes derived from the same parent. 
Example
Imagine a Vehicle class with attributes like manufacturer and model.
In a hierarchical structure, a Car class and a Motorcycle class could both inherit from Vehicle.
Both Car and Motorcycle would automatically have the manufacturer and model properties from Vehicle.
However, Car could have a unique method like openTrunk(), while Motorcycle could have startEngine(). 


```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def intro(self):
        print(f"My name is {self.name}")
```

```python
class Cat(Animal):
    def speak(self):
        return "Meow"
    
class dog(Animal):
    def speak(self):
        return "woof"
    
class Lion(Animal):
    def speak(self):
        return "roar"
```

###### Hybrid Inheritance

Hyybrid inheritance is a combination of two or more types of inheritance from the above types.

```python
class Animal:
    def __init__(self, name, age =2):
        self.name = name
        self.age = age
        
    def intro(self):
        print(f"My name is {self.name}")
```

```python
class Mammal(Animal):
    def __init__(self, name, fur = False):
        super(). __init__(name)
        self.fur = fur
        
    def birth(self):
        print(f"{self.name} give birth to live ones")
        
class Birds(Animal):
    def fly(self):
        print(f"{self.name} can fly")
```

```python
class Bat(Mammal, Birds):
    pass
```

```python
b = Bat("Bruce", 4)
```

```python
b.intro()
```

```python
b.fly()
```

```python
b.birth()
```
