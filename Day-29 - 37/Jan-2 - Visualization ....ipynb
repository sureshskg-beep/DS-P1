## 7. Heat Map

```python
correlation = titabic_df.corr(numeric_only = True)
```

```python
correlation
```

```python
sns.heatmap(data = correlation, cmap = "coolwarm")
```

```python
sns.heatmap(data = correlation, annot = True, cmap = "coolwarm")
```

```python
sns.heatmap(data = correlation, annot = True, fmt = ".2f", cmap = "coolwarm")
```

```python
sns.heatmap(data = correlation, annot = True, fmt = ".2f", cmap = "jet")
```

```python
import numpy as np
mask1 = np.triu(np.ones_like(correlation, dtype=bool))
```

```python
mask1
```

```python
mask2 = np.tril(np.ones_like(correlation, dtype=bool))
```

```python
mask2
```

```python
sns.heatmap(data = correlation, annot = True, fmt = ".2f", cmap = "jet", mask=mask2)
```

```python
sns.heatmap(data = correlation, annot = True, fmt = ".2f", cmap = "jet", mask=mask1)
```

# 8. Density Plot

Density plot is similar to 3. histograph

```python
titabic_df.Age.plot(
    kind = "kde",
    grid = True,
    ind = 20
)
```

```python
titabic_df.Age.plot(
    kind = "kde",
    grid = True,
    ind = np.linspace(0, 500, 5000),
    figsize = (15, 5)
)
```

```python
titabic_df.Survived.plot(
    kind = "kde",
    grid = True,
    ind = np.linspace(0, 25, 50),
    figsize = (15, 5)
)
```
