# Normalization - data handling 

```python
titanic_normalization = pd.read_csv("final_data_titanic.csv")
```

```python
titanic_normalization.head()
```

```python
min_fare = titanic_normalization.Fare.min()
max_fare = titanic_normalization.Fare.max()
```

```python
norm_fare = ( titanic_normalization.Fare - min_fare) / (max_fare - min_fare)
```

```python
norm_fare
```

```python
norm_fare.describe()
```

```python
titanic_normalization.Fare.describe()
```

```python
norm_fare.plot(kind = "kde")
```

```python
titanic_normalization.Fare.plot(kind = "kde")
```
