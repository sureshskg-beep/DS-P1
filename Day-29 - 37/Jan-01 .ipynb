```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("Titanic_data.csv")
```

```python
titanic_df["Survived"].value_counts().plot(
    kind = "bar",
    title = "count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired" 
    
)
```

```python
titanic_df["Survived"].value_counts().plot(
    kind = "bar",
    title = "count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired" 
    
)
```

```python
pd.crosstab(
    titanic_df.Survived,
    titanic_df.Sex
).plot(
    kind = "bar",
    title = "count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "Paired" 
)
```

## 2. Pie Chart

```python
group_by_embarked = titanic_df.groupby(by = "Embarked")
```

```python
group_by_embarked.count()
```

```python
group_by_embarked.count().plot(
    kind = "pie", y = "Survived"
)
```

```python
group_by_embarked.count().plot(
    kind = "pie", subplots = True
)
```

```python
group_by_embarked.count().plot(
    kind = "bar",
    title = "count of Survived Passengers",
    figsize = (4,3),
    legend = True,
    xlabel = "survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    yticks = list(range(0,600,70)),
    colormap = "tab20" 
)
```

```python
group_by_embarked["Survived"].value_counts().plot(
    kind="pie",
    title="Count of Survived Passengers",
    figsize=(6, 5),
    autopct="%1.1f%%",
    startangle=90,
    colormap="tab20"
)
```

```python
import matplotlib.pyplot as plt

titanic_df["Age"].sort_values().plot(
    kind="line",
    figsize=(15,4)
)

plt.title("Passenger Age Trend (Sorted)")
plt.xlabel("Passenger Index (Sorted by Age)")
plt.ylabel("Age")
plt.show()
```

## 5. Scatter Plot

```python
titanic_df.plot(
    kind = "scatter",
    x = "Age",
    y = "Fare"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(
    titanic_df["Age"],
    titanic_df["Fare"]
)

plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
titanic_df.plot(
    kind = "scatter",
    x = "Age",
    y = "Fare",
    c = "Survived",
    colormap = "tab20"
)
```

```python
titanic_df.plot(
    kind = "scatter",
    x = "Age",
    y = "Fare",
    c = "SibSp",
    colormap = "bwr"
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    data=titanic_df,
    x="Pclass",
    hue="Survived"
)

plt.title("Passenger Count by Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()
```
