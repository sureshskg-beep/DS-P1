```python
class CashLimitException(Exception):
    """Raised when the cash balance is less than 5000"""
    pass


class BankAccount:
    def __init__(self, total_value=3000):
        self.total_value = total_value

    def get_cash(self, withdraw):
        if self.total_value - withdraw >= 5000:
            self.total_value -= withdraw
            print("Please collect the cash")
            print("Remaining balance:", self.total_value)
        else:
            raise CashLimitException("Not enough balance in your account")
```

```python
account = BankAccount(total_value=10000)

try:
    account.get_cash(8000)
except CashLimitException as e:
    print(e)
```
