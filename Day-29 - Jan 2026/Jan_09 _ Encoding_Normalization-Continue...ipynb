# log transformation

```python
import numpy as np

# Parameters
n = 500                 # points for growth and decay
start = 1.0
peak = 5 * 10e3        # 50,000

# Exponential growth
growth = np.exp(np.linspace(np.log(start), np.log(peak), n))

# Exponential decay back to 1
decay = np.exp(np.linspace(np.log(peak), np.log(start), n))

# Combine (remove duplicate peak)
data = np.concatenate([growth, decay[1:]])

```

```python
pd.DataFrame(data, columns=["data"])
```

```python
pd.DataFrame(data, columns=["data"]).plot(kind="line")
```

```python
pd.DataFrame(np.log(data), columns=["data"])
```

```python
pd.DataFrame(np.log(data), columns=["data"]).plot(kind = 'line')
```

```python
def boxcox_transform(x, lam):
    x = np.asarray(x, dtype=float)
    if np.any(x <= 0):
        raise ValueError("Box–Cox requires strictly positive values.")
    if lam == 0:
        return np.log(x)
    return (np.power(x, lam) - 1) / lam

x_boxcox = boxcox_transform(x, lam=0.3)
x_boxcox
```

```python
def yeojohnson_transform(x, lam):
    x = np.asarray(x, dtype=float)
    out = np.zeros_like(x)

    pos = x >= 0
    neg = ~pos

    if lam != 0:
        out[pos] = ((x[pos] + 1) ** lam - 1) / lam
    else:
        out[pos] = np.log(x[pos] + 1)

    if lam != 2:
        out[neg] = -(((-x[neg] + 1) ** (2 - lam) - 1) / (2 - lam))
    else:
        out[neg] = -np.log(-x[neg] + 1)

    return out

x_yj = yeojohnson_transform(x, lam=0.5)
x_yj
```
