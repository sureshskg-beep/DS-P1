# Inerting and Deleting Column and Rows

```python
titanic_df.drop(["PassengerId"], axis = 1, inplace= False)
```

```python
# Dropping a Row:

titanic_df.drop([0,2,4], axis =0).head()
```

```python
initials = [name.split(",")[-1].split(".")[0] for name in titanic_df["Name"]]
len(initials)
```

```python
titanic_df["initials"] = initials
```

```python
titanic_df.head()
```

```python
titanic_df.insert(loc=1, column="new_initial", value=initials)
```

## Handling Null or Nan or NA values


1. Removing Null vlues
2. Fill Something (imputation)
3. Null as new feature (Feature Engineering)


```python
titanic_df.isna().sum()
```

```python
df_filter_1 = titanic_df.dropna(subset = "Embarked", axis=0, ignore_index=True) # delete dropna
```

```python
df_filter_1
```

```python
df_filter_1.isna().sum()
```

```python
titanic_df.Cabin.notna()
```

```python
df_filter_1 ['Cabin'] = titanic_df.Cabin.notna().astype(int)   #replace
```

```python
df_filter_1.isna().sum()
```

```python
# fillup
print(titanic_df.Age.mean())
titanic_df.Age.plot(kind = "hist", bins = 20)
```

```python
df_filter_1["Age"] = titanic_df.Age.fillna
```
