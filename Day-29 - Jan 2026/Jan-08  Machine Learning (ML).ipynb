```python
import pandas as pd

cleaned_titanic_df = pd.read_csv("Cleaned_Titanic.csv")
```

```python
cleaned_titanic_df.head()
```

```python
cleaned_titanic_df = cleaned_titanic_df.drop(
    columns=["Age", "Name", "Ticket", "SibSp", "Pclass"]
)
```

```python
cleaned_titanic_df = cleaned_titanic_df.drop(
    columns=["Unnamed: 0", "initials", "new_initial"]
)
```

```python
cleaned_titanic_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False)
```

```python
one_hot_encoder
```

```python
encoder_gender = one_hot_encoder.fit_transform(cleaned_titanic_df[["Sex"]])
```

```python
encoder_gender
```

```python
cleaned_titanic_df["Sex"]
```

```python
one_hot_encoder.categories_
```

```python
one_hot_encoder.inverse_transform(
    [[1,0],
     [0,1]]
)
```

```python
one_hot_encoder.inverse_transform(
    encoder_gender
)
```

```python
df = pd.DataFrame(encoder_gender, columns=one_hot_encoder.categories_[0])
```

```python
cleaned_titanic_df = pd.concat([cleaned_titanic_df, df], axis = 1)
cleaned_titanic_df
```
