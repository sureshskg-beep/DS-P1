```python
import pandas as pd

data = pd.read_csv("Iris.csv", index_col=0)
data.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
X = data.drop("Species", axis=1).to_numpy()
y = data[["Species"]].to_numpy()
```

```python
scaler = StandardScaler()
onehot_encoder = OneHotEncoder(sparse_output=False)

enc_y = onehot_encoder.fit_transform(y)
scaled_X = scaler.fit_transform(X)
```

```python
cat_array = onehot_encoder.categories_[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    scaled_X, enc_y, test_size=0.2, shuffle=True, stratify=enc_y, random_state=999
)
```

```python
X_train.shape
```

```python
y_train.shape
```

# Model Creation and Compile 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python
inp_shape = X_train.shape[1]
num_clases = y_train.shape[1]
```

```python
model = Sequential()

model.add(Input(shape= (inp_shape,)))

model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

model.add(Dense(num_clases, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(
    optimizer="Adam", loss="categorical_crossentropy", metrics=["accuracy", "f1_score"]
)
```

```python
history = model.fit(
    X_train, y_train, epochs=25, batch_size=32, verbose=1
)
```

```python
history.history
```

```python
model.evaluate(X_test, y_test)

```

```python
y_pred = np.argmax(model.predict(X_test), axis=1)
```

```python
y_true = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_true, y_pred)
```

```python
accuracy_score(y_true, y_pred)
```

```python
f1_score(y_true, y_pred, average="macro")
```

```python
precision_score(y_true, y_pred, average="macro")
```

```python
recall_score(y_true, y_pred, average="macro")
```
