```python
import pickle
import numpy as np
import tensorflow as tf
```

```python
with open("iris_classifier.pkl", "rb") as file:
    objects  = pickle.load(file)
```

```python
scaler = objects['scaler']
encoder = objects['encoder']
model = objects['model']
```

```python


model = tf.keras.models.load_model("iris_classification.keras")
```

```python
#user Input

my_data = np.array([
    [12.6, 2.6, 8.5, 0.9],
    [6.1, 3.0, 4.9, 1.8]
])
```

```python
# pre Processing 

scaled_data = scaler.transform(my_data)

```

```python
scaled_data
```

```python
output = np.argmax(model.predict(scaled_data), axis=1)
```

```python
cat = encoder.categories_[0]
values = [cat[res] for res in output]
values 
```
